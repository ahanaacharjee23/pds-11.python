```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns


# Step 1: Create term-test-1-result.csv
np.random.seed(42)
registration_numbers = [f'REG{i:03d}' for i in range(1, 51)]
names = [f'Student{i}' for i in range(1, 51)]
tt1_marks = np.random.randint(0, 21, size=50)  # Random marks between 0 and 20

df1 = pd.DataFrame({
    'Registration Number': registration_numbers,
    'Name': names,
    'TT-1 Marks': tt1_marks
})
df1.to_csv('term-test-1-result.csv', index=False)
df1
```

```python
# Step 2: Create term-test-2-result.csv
tt2_marks = np.random.randint(0, 21, size=50)

df2 = pd.DataFrame({
    'Registration Number': registration_numbers,
    'Name': names,
    'TT-2 Marks': tt2_marks
})
df2.to_csv('term-test-2-result.csv', index=False)
df2
```

```python
# Step 3: Load and merge DataFrames
df1 = pd.read_csv('term-test-1-result.csv')
df2 = pd.read_csv('term-test-2-result.csv')
merged_df = pd.merge(df1, df2, on='Registration Number')
merged_df
```

```python
# Step 4: Add Best and Average Marks columns
merged_df['Best Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)
merged_df['Average Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)
merged_df
```

```python
# Step 5: Drop unnecessary columns
merged_df = merged_df.drop(columns=['TT-1 Marks', 'TT-2 Marks'])
merged_df
```

```python
# Step 6: Create attendance and term final marks CSV
attendance_marks = np.random.randint(0, 11, size=50)  # Random marks between 0 and 10
term_final_marks = np.random.randint(0, 101, size=50)  # Random marks between 0 and 100

attendance_df = pd.DataFrame({
    'Registration Number': registration_numbers,
    'Attendance Marks': attendance_marks,
    'Term Final Marks': term_final_marks
})
attendance_df.to_csv('attendance-term-final.csv', index=False)
attendance_df
```

```python
# Step 7: Merge and calculate Final Marks
attendance_df = pd.read_csv('attendance-term-final.csv')
final_df = pd.merge(merged_df, attendance_df, on='Registration Number')
final_df['Final Marks'] = (final_df['Term Final Marks'] * 0.7) + final_df['Average Marks'] + final_df['Attendance Marks']
final_df.to_csv('final-result.csv', index=False)
final_df
```

```python
# Step 8: K-Means Clustering and Visualization
print("Step 8: Performing K-Means clustering on Final Marks...")
X = final_df[['Final Marks']]
inertia = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 6), inertia, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()

optimal_k = 3  # Based on the elbow method, adjust if needed
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
final_df['Cluster'] = kmeans.fit_predict(X)

print("\nClustered Data (Sample):")
print(final_df[['Registration Number', 'Final Marks', 'Cluster']].head())

plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=final_df['Registration Number'],
    y=final_df['Final Marks'],
    hue=final_df['Cluster'],
    palette='viridis',
    s=100
)
plt.title('K-Means Clustering of Final Marks')
plt.xlabel('Registration Number')
plt.ylabel('Final Marks')
plt.xticks(rotation=90)
plt.legend(title='Cluster')
plt.show()

final_df.to_csv('clustered-final-result.csv', index=False)
print("Clustered results saved to clustered-final-result.csv!\n")
```
